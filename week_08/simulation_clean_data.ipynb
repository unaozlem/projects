```python
df_mon = pd.read_csv("./data/monday.csv", sep = ';', index_col=False, parse_dates=['timestamp'])
df_tu = pd.read_csv("./data/tuesday.csv", sep = ';', index_col=False, parse_dates=['timestamp'])
df_wed = pd.read_csv("./data/wednesday.csv", sep = ';', index_col=False, parse_dates=['timestamp'])
df_th = pd.read_csv("./data/thursday.csv", sep = ';', index_col=False, parse_dates=['timestamp'])
df_fr = pd.read_csv("./data/friday.csv", sep = ';', index_col=False, parse_dates=['timestamp'])
```

```python
df_all_1 = pd.concat([df_mon, df_tu], ignore_index=True)
df_all_1.tail()
```

```python
df_all_2 = pd.concat([df_all_1, df_wed], ignore_index=True)
df_all_2.tail()
```

```python
df_all_3 = pd.concat([df_all_2, df_th], ignore_index=True)
df_all_3.tail()
```

```python
df_all = pd.concat([df_all_3, df_fr])
df_all.tail()
```

```python
df_new = df_all.copy().sort_values("timestamp")
```

```python
df_ent =df_new.groupby('customer_no').first().reset_index()
df_ent
```

```python
df_ent["timestamp"] = df_ent["timestamp"] - pd.Timedelta(minutes=1) 
df_ent
```

```python
df_ent["location"]="entrance"
df_ent
```

```python
# missing values of enterence added 
#(one minute earlier than the time that they have been seen for the first time.)
df_all_ent = pd.concat([df_all, df_ent])
```

```python
df_check = df_all_ent.copy().sort_values("timestamp")
```

```python
df_check = df_new.groupby('customer_no').last().reset_index()
df_check
```

```python
df_check = df_check[df_check["location"] != "checkout"]
df_check
```

```python
df_check["timestamp"] = df_check["timestamp"] + pd.Timedelta(minutes=1) 
df_check
```

```python
df_check["location"]="checkout"
df_check
```

```python
df_clean = pd.concat([df_all_ent, df_check]).sort_values("timestamp").reset_index()
df_clean
```

```python
df_clean[df_clean["customer_no"] == 10006]
```

```python
df_clean = (
    df_clean.set_index("timestamp")
    .groupby("customer_no")[["location"]]
    .resample(rule="1min")
    .ffill()
    .reset_index()
)
```

```python
df_clean.to_csv("data/simulation_data_clean.csv")
```
