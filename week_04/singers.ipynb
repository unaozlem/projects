```python
import requests
import re
```

**The Rolling Stones**

```python
response_1 = requests.get('https://www.lyrics.com/lyric/36144607/The+Rolling+Stones/Start+Me+Up')
rolling_stones_html = response_1.text
```

```python
response_1.status_code
```

```python
with open ("rolling_stones_html", "w") as r:
    r.write(rolling_stones_html)
```

```python
print(response_1.text)
```

```python
roll_links = re.findall('href.{100}', response_1.text)
roll_links
```

```python
# You will automate this part for your project
song_links = ['/The+Rolling+Stones/Rough+Justice',
              '/The+Rolling+Stones/Hate+to+See+You+Go',
              '/The+Rolling+Stones/One+More+Shot']
```

```python
for counter, link in enumerate(song_links):
    url = 'https://www.lyrics.com/lyric/36144607/The+Rolling+Stones/Start+Me+Up' + link
    response_roll = requests.get(url)
    with open(f'{counter}.txt', 'w', encoding='utf-8') as f:
        f.write(response_roll.text)
    print(counter, link, response_roll.status_code)
```
